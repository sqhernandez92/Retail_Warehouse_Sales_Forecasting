```python
import pandas as pd

df = pd.read_csv("/Users/stephaniehernandez/Desktop/WarehouseandRetailSales/Warehouse_and_Retail_Sales.csv")
df.head()
```

```python
df.columns = df.columns.str.strip().str.replace(' ', '_')

df['DATE'] = pd.to_datetime(df['YEAR'].astype(str) + '-' + df['MONTH'].astype(str) + '-01')

df = df.sort_values('DATE')
```

```python
monthly_sales = df.groupby('DATE')[['RETAIL_SALES', 'WAREHOUSE_SALES']].sum()
monthly_sales.head()
```

```python
from prophet import Prophet
import matplotlib.pyplot as plt

retail_df = monthly_sales[['RETAIL_SALES']].reset_index()
retail_df.columns = ['ds', 'y']

model = Prophet()
model.fit(retail_df)

future = model.make_future_dataframe(periods=12, freq='M')
forecast = model.predict(future)

model.plot(forecast)
plt.title('Retail Sales Forecast')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

```python
from prophet import Prophet
import matplotlib.pyplot as plt


retail_df = monthly_sales[['RETAIL_SALES']].reset_index()
retail_df.columns = ['ds', 'y']


retail_model = Prophet()
retail_model.fit(retail_df)


retail_future = retail_model.make_future_dataframe(periods=12, freq='M')
retail_forecast = retail_model.predict(retail_future)


warehouse_df = monthly_sales[['WAREHOUSE_SALES']].reset_index()
warehouse_df.columns = ['ds', 'y']


warehouse_model = Prophet()
warehouse_model.fit(warehouse_df)


warehouse_future = warehouse_model.make_future_dataframe(periods=12, freq='M')
warehouse_forecast = warehouse_model.predict(warehouse_future)


plt.figure(figsize=(12,6))
plt.plot(retail_forecast['ds'], retail_forecast['yhat'], label='Retail Forecast', color='blue')
plt.plot(warehouse_forecast['ds'], warehouse_forecast['yhat'], label='Warehouse Forecast', color='orange')
plt.title('Forecast Comparison: Retail vs Warehouse Sales')
plt.xlabel('Date')
plt.ylabel('Predicted Sales')
plt.legend()
plt.show()
```
